```
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

cap = cv2.VideoCapture("/content/drive/MyDrive/Mini-MLProject-main/TestVedio.mp4")

# Detecting a car using the car cascade 
car_cascade = cv2.CascadeClassifier("/content/drive/MyDrive/Mini-MLProject-main/cars.xml")
 
c = 1
while True:
  
    ret, frames = cap.read()
    # 'ret' is the status of reading the picture
    # 'frame' is the picture of the current frame
    
    if not ret: 
      print("Video is completed")
      break
    # rgb 2 grey
    grey = cv2.cvtColor(frames, cv2.COLOR_BGR2GRAY)
    #Gaussian blur
    blur = cv2.GaussianBlur(grey,(5,5),0)
    #dilate
    dilated = cv2.dilate(blur,np.ones((3,3)))
    # Morphology-Ex, Structural Elements
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2, 2))
    closing = cv2.morphologyEx(dilated, cv2.MORPH_CLOSE, kernel) 

    cars = car_cascade.detectMultiScale(closing, 1.1, 1)

    cnt = 0
    for (x, y, w, h) in cars:
        cv2.rectangle(frames, (x, y), (x + w, y + h), (0, 0, 255), 2)
 
        cnt += 1

    if c <= 2000:
        if c % 10 == 0:
            cv2_imshow(frames)
            success, image = cap.read()
            resize = cv2.resize(frames, (340, 220), interpolation=cv2.INTER_LINEAR)
            cv2.imwrite("%03ddetection.jpg" % c, resize)
        c += 1
 
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            print("Video detection halted.")
            break
    else:
        print("Video is completed.")
        break
    print(cnt, " cars found")
cap.release()
cv2.destroyAllWindows()
```
